```python
import pandas as pd
from google.colab import drive
from os.path import join

drive.mount('./drive', force_remount=True)
path_prefix = './drive/My Drive'

from google.colab import drive
drive.mount('/content/drive')

```

```python
fname='consurf.csv' #excelde dosyanızı kaydettiğiniz ad
df = pd.read_csv(join(path_prefix, fname))
df
```

```python
df['COLOR']= df.COLOR.astype(int)
```

```python
Conserved_Amino_acids=pd.DataFrame(columns=['Amino Acids','Position','Color'])
Pos=[]
Seq=[]
Color=[]
copy=df.copy()
for i in df.index:
  if(7<=copy['COLOR'][i]<=9):
    Pos.append(copy['POS'][i])
    Seq.append(copy['SEQ'][i])
    Color.append(copy['COLOR'][i])

Conserved_Amino_acids['Amino Acids']=Seq
Conserved_Amino_acids['Position']=Pos
Conserved_Amino_acids['Color']=Color

Conserved_Amino_acids.sort_values(by='Position')
```

```python
Conserved_Amino_acids['Color'].describe()
```

```python
Position=[]
Sequence=[]
ColorNum=[]
for i in df.index:
  if(1<=df['COLOR'][i]<=6):
   Position.append(df['POS'][i])
   ColorNum.append(df['COLOR'][i])
   Sequence.append(df['SEQ'][i])

Unconserved_Amino_acids=pd.DataFrame(columns=['Amino Acids','Position','Color'])

Unconserved_Amino_acids['Amino Acids']= Sequence
Unconserved_Amino_acids['Position']=Position
Unconserved_Amino_acids['Color']=ColorNum
Unconserved_Amino_acids
```

```python
aa_names = {'CYS': 'C', 'ASP': 'D', 'SER': 'S', 'GLN': 'Q', 'LYS': 'K',
         'ILE': 'I', 'PRO': 'P', 'THR': 'T', 'PHE': 'F', 'ASN': 'N', 
         'GLY': 'G', 'HIS': 'H', 'HSD': 'H', 'LEU': 'L', 'ARG': 'R', 'TRP': 'W', 
         'ALA': 'A', 'VAL':'V', 'GLU': 'E', 'TYR': 'Y', 'MET': 'M'}
mutation_list = ["ALA","ARG","ASN","ASP","CYS","GLU","GLN","GLY","HIS","ILE",
                 "LEU","LYS","MET","PHE","PRO","SER","THR","TRP","TYR","VAL"]
```

```python
# mutation on %5 of conserved amino acids
import random as rnd
check=1
M_Conserved=Conserved_Amino_acids.copy()
row,col=Conserved_Amino_acids.shape
while(check<263*5/100):  #263 ü kendi proteninizin uzunluğu ile değiştirirn
  ranom_index=rnd.randint(0,row-1)
  random_mut=aa_names[mutation_list[rnd.randint(0,len(mutation_list)-1)]]

  if(random_mut!=M_Conserved.at[ranom_index,'Amino Acids']):
    print( M_Conserved.at[ranom_index,'Amino Acids']+"--->"+random_mut+" mutation at "+str(ranom_index))
    M_Conserved.at[ranom_index,'Amino Acids']=random_mut
  
    check+=1

M_Conserved 
```

```python
# mutation on %5 of unconserved amino acids
check=1
M_Unconserved=Unconserved_Amino_acids.copy()
row,col=Unconserved_Amino_acids.shape
while(check<263*5/100):#263 ü kendi proteninizin uzunluğu ile değiştirirn
  ranom_index=rnd.randint(0,row-1)
  random_mut=aa_names[mutation_list[rnd.randint(0,len(mutation_list)-1)]]

  if(random_mut!=M_Unconserved.at[ranom_index,'Amino Acids']):
    print( M_Unconserved.at[ranom_index,'Amino Acids']+"--->"+random_mut+" mutation at "+str(ranom_index))
    M_Unconserved.at[ranom_index,'Amino Acids']=random_mut
  
    check+=1

M_Unconserved 
```

```python
# mutation on %5 of conserved and unconserved amino acids
frmes= [ M_Conserved,M_Unconserved]
x=pd.concat(frmes,ignore_index=True)
x.sort_values(by='Position')
mutated_Seq="" 
for i in x['Amino Acids'].values:
  mutated_Seq+=i
mutated_Seq
```

```python
#mutation on %10 of  the whole protein
check=1
mut=""
seq=df.copy()
row,col=Unconserved_Amino_acids.shape
while(check<263*10/100):
  ranom_index=rnd.randint(0,row-1)
  random_mut=aa_names[mutation_list[rnd.randint(0,len(mutation_list)-1)]]

  if(random_mut!=seq.at[ranom_index,'SEQ']):
    print( seq.at[ranom_index,'SEQ']+"--->"+random_mut+" mutation at "+str(ranom_index))
    seq.at[ranom_index,'SEQ']=random_mut
  
    check+=1
```
